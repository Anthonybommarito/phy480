```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt('derivative_test.dat')
```

```python
plt.scatter(data[0:,0],data[0:,1])
def f(x,m,b):
    return m*x+b

popt, pcov = curve_fit(f, data[0:,0], data[0:,1])
plt.plot(data[0:,0], f(data[0:,0], *popt), label= 'y=-0.81167222x-14.36888881')
plt.legend()
plt.title('error plot of diff_extrap2')
popt
plt.ylabel(r'$\log (error)$',fontsize=12)
plt.xlabel(r'$\log (h)$',fontsize=12)
```

```python
x=data[0:,0]
x_clean=x[0:3]
y=data[0:,1]
y_clean=y[0:3]
```

```python
plt.scatter(x_clean,y_clean)
popt, pcov = curve_fit(f, x_clean, y_clean)
plt.plot(x_clean, f(x_clean, *popt), label= 'y=6.01007081x -7.31036966')
plt.title('error plot of diff_extrap2 clean fit')
plt.ylabel(r'$\log (error)$',fontsize=12)
plt.xlabel(r'$\log (h)$',fontsize=12)
plt.legend()
```
